### Description 
> #### In this game, the user gets the first chance to pick the option among Rock, paper and scissors. After that computer select from the remaining two choices(randomly), the winner is decided as per the rules.
- Rock vs paper-> paper wins
- Rock vs scissor-> Rock wins
- paper vs scissor-> scissor wins.

```python
import random
import matplotlib.pyplot as plt

def display_image(choice):
    if choice == 0:
        rock = plt.imread("rock.png")
        plt.imshow(rock)
        plt.title("Rock")
        plt.show()
    elif choice == 1:
        paper = plt.imread("paper.png")
        plt.imshow(paper)
        plt.title("Paper")
        plt.show()
    elif choice == 2:
        scissors = plt.imread("images.png")
        plt.imshow(scissors)
        plt.title("scissors")
        plt.show()
        return

def play_game():
    print("Welcome to rock paper scissors game! Let's play! 🎮🕹️👾")
    user_choice = input("Enter your choice (rock, paper, or scissors): ").lower()
    if user_choice == "rock":
        user_choice_num = 0
    elif user_choice == "paper":
        user_choice_num = 1
    elif user_choice == "scissors":
        user_choice_num = 2
    else:
        print("Invalid choice. Please enter rock, paper, or scissors.")
        return

    computer_choice = random.randint(0, 2)
    display_image(computer_choice)

    if user_choice_num >= 3 or user_choice_num < 0:
        print("Invalid number please add numbers 0, 1, or 2")
    elif computer_choice == 0 and user_choice_num == 2:
        print("Sorry 😔\nComputer wins")
    elif (computer_choice == 1 and user_choice_num == 0) or (computer_choice == 2 and user_choice_num == 1):
        print("Sorry 😔\nComputer wins")
    elif computer_choice > user_choice_num:
        print("Congratulations 😊 You win")
    elif computer_choice < user_choice_num:
        print("Congratulations 😊 You win")
    else:
        print("Try again")
        return
play_game()
```
